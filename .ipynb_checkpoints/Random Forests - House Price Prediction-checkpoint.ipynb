### 1. Objective
The goal of this project is to predict the selling price of homes (`SalePrice`) based on various features such as the lot size, year built, and number of rooms. 

We will use a **Random Forest Regressor**, a powerful ensemble learning method that operates by constructing a multitude of decision trees during training. It is chosen for its ability to handle non-linear relationships and resistance to overfitting compared to single decision trees.

### 2. Dataset Overview
The dataset contains housing data from Ames, Iowa.
* **Rows:** 1460
* **Columns:** 81
* **Target Variable:** `SalePrice`

```python
# 1. Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.max_columns', None)

print("Libraries imported successfully!")
```

### 3. Load and Inspect Data
We will load the dataset and look at the first few rows to understand the structure. We also need to check for data types and missing values.

```python
# Load the dataset
file_path = './home-data-for-ml-course/train.csv'
df = pd.read_csv(file_path)

# Display shape and first 5 rows
print(f"Dataset Shape: {df.shape}")
df.head()
```

```python
# Check for data types and missing values
df.info()
```

### 4. Data Preprocessing
We need to select relevant features and handle any missing data. Random Forests generally do not support missing values in standard sklearn implementations.

**Selected Features:**
* `LotArea`: Lot size in square feet
* `YearBuilt`: Original construction date
* `1stFlrSF`: First Floor square feet
* `2ndFlrSF`: Second floor square feet
* `FullBath`: Full bathrooms above grade
* `BedroomAbvGr`: Bedrooms above grade (does not include basement bedrooms)
* `TotRmsAbvGrd`: Total rooms above grade (does not include bathrooms)
* `OverallQual`: Rates the overall material and finish of the house (Added this as it is usually a strong predictor)

```python
# Define Features and Target
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 
            'BedroomAbvGr', 'TotRmsAbvGrd', 'OverallQual']
target = 'SalePrice'

X = df[features]
y = df[target]

# Check for missing values in our selected features
print("Missing values before imputation:\n")
print(X.isnull().sum())

# Impute missing values (if any) with the median
imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

print("\nData preprocessing complete.")
```

### 5. Exploratory Data Analysis (EDA)
Before modeling, we must understand the distribution of our target variable and how features correlate with it.

#### 5.1 Target Variable Distribution
We check if the `SalePrice` follows a normal distribution or if it is skewed.

```python
# Visualization: Distribution of Sale Price
plt.figure(figsize=(10, 6))
sns.histplot(y, kde=True, color='blue', bins=30)
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price ($)')
plt.ylabel('Frequency')
plt.axvline(y.mean(), color='red', linestyle='--', label=f'Mean Price: ${y.mean():,.0f}')
plt.legend()
plt.show()
```

### Analysis: Target Variable Distribution
**What the Graph Shows:**
The histogram shows that `SalePrice` is **right-skewed** (positively skewed). Most houses are sold between \$100k and \$200k, with a long "tail" extending towards expensive luxury homes (\$500k+).

**Implication for Random Forest:**
* **Non-Parametric Nature:** Unlike Linear Regression, Random Forest does **not** assume the target variable follows a Normal (Bell curve) distribution. It can handle this skewness reasonably well.
* **Prediction Limitation:** However, tree-based models generally cannot "extrapolate." If the most expensive house in our training data is \$750k, the model will likely never predict a price higher than that for a new, superior house. The skewness suggests we have fewer examples of high-end homes, making the model potentially less accurate for luxury properties.

#### 5.2 Correlation Matrix
A heatmap helps us identify which features are most strongly correlated with the `SalePrice`. Darker colors indicate stronger relationships.

```python
# Correlation Matrix
plt.figure(figsize=(10, 8))
# Combine X and y temporarily for correlation analysis
analysis_df = X.copy()
analysis_df['SalePrice'] = y

corr_matrix = analysis_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

### Analysis: Correlation Matrix
**What the Graph Shows:**
The heatmap visualizes linear relationships.
* **Red Squares (High Correlation):** `OverallQual` (0.79) and `1stFlrSF` (0.61) have the strongest positive correlation with `SalePrice`.
* **Blue/Light Squares:** Features like `LotArea` have a weaker linear correlation.

**Implication for Random Forest:**
* **Non-Linearity:** While this heatmap shows *linear* ties, Random Forest excels at finding **non-linear** relationships that this map might miss. For example, `LotArea` might only matter significantly if the `Neighborhood` is wealthy. A simple correlation number won't show that, but the Random Forest will learn it.
* **Multicollinearity:** Notice that `1stFlrSF` and `TotalBsmtSF` might be correlated with each other. In Linear Regression, this breaks the model. In Random Forest, it distributes the "importance" score between them, making it safer to use correlated features.

### 6. Model Training
We will split the data into training and validation sets to ensure we can evaluate the model on unseen data.

* **Model:** Random Forest Regressor
* **Split:** 80% Training, 20% Validation
* **Metric:** Mean Absolute Error (MAE)

```python
# Split data
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state=1)

# Initialize the model
# n_estimators=100 means we create 100 decision trees
rf_model = RandomForestRegressor(n_estimators=100, random_state=1)

# Fit the model
print("Training model...")
rf_model.fit(train_X, train_y)
print("Model training complete.")
```

### 7. Model Evaluation
We evaluate the model using **MAE** (Mean Absolute Error), which tells us the average dollar amount our prediction is off by, and **$R^2$**, which tells us how well the model explains the variance in the data.

```python
# Make predictions
val_predictions = rf_model.predict(val_X)

# Calculate metrics
mae = mean_absolute_error(val_y, val_predictions)
rmse = np.sqrt(mean_squared_error(val_y, val_predictions))
r2 = r2_score(val_y, val_predictions)

print("Model Performance Metrics:")
print("="*30)
print(f"Mean Absolute Error (MAE): ${mae:,.2f}")
print(f"Root Mean Sq Error (RMSE): ${rmse:,.2f}")
print(f"R-squared Score (R^2):     {r2:.4f}")
```

### Analysis: Evaluation Metrics
**The Numbers:**
* **$R^2 = 0.89$:** Our model explains **89% of the variation** in house prices. This is a strong score for a first iteration.
* **MAE = \$19,746:** On average, our prediction is off by about \$20k.

**Implication for Random Forest:**
* If the MAE is acceptable for the business context (e.g., a real estate estimator), the model is usable.
* However, Random Forests can be prone to **overfitting**. If your Training Score is 0.99 and Validation Score is 0.89, the model is "memorizing" the data rather than learning the rules. (We should check training score in the future to confirm this).

#### 7.1 Actual vs. Predicted Plot
This scatter plot allows us to visually inspect how close our predicted prices are to the actual prices. Ideally, all points should lie on the red diagonal line.

```python
# Visualization: Actual vs Predicted
plt.figure(figsize=(10, 6))
sns.scatterplot(x=val_y, y=val_predictions, alpha=0.6)
plt.plot([val_y.min(), val_y.max()], [val_y.min(), val_y.max()], 'r--', lw=2) # Diagonal line
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.title('Actual vs. Predicted Prices (Random Forest)')
plt.tight_layout()
plt.show()
```

### Analysis: Actual vs. Predicted
**What the Graph Shows:**
* **The Diagonal:** The red dashed line represents a perfect prediction.
* **The Spread:** The blue dots cluster tightly around the line for prices up to \$300k, indicating high accuracy for average homes.
* **The Divergence:** As prices go higher (right side of x-axis), the dots start to drift below the red line.

**Implication for Random Forest:**
* **Underestimation of High Values:** The model tends to **under-predict** expensive houses. This is a classic Random Forest characteristic in regression; it averages the values of leaf nodes, tending to pull extreme outliers back toward the mean.
* **Outliers:** The stray points far from the line represent properties with unique characteristics (e.g., a historic mansion) that our selected features (size, room count) failed to capture.

#### 7.2 Feature Importance
One of the advantages of Random Forest is that it provides feature importance scores. This tells us which factors drive the price the most.

```python
# Get feature importances
importances = rf_model.feature_importances_
feature_names = X.columns

# Create a DataFrame for visualization
feature_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_imp_df = feature_imp_df.sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_imp_df, palette='viridis')
plt.title('Random Forest Feature Importance')
plt.xlabel('Importance Score')
plt.show()
```

### Analysis: Feature Importance
**What the Graph Shows:**
This plot displays how much each feature contributes to reducing prediction error across all trees in the forest.

**Key Insights:**
1.  **OverallQual is King:** The overall material and finish of the house is by far the most critical predictor.
2.  **Size vs. Count:** Square footage (`1stFlrSF`, `2ndFlrSF`) is more important than the mere count of rooms (`TotRmsAbvGrd`). A small house with many tiny rooms is worth less than a spacious open-plan home.
3.  **Actionable Insight:** For a homeowner looking to increase value, improving the *quality* (renovations) is likely more effective than building an addition (increasing `LotArea`), as `LotArea` has low importance in this specific model configuration.

### 8. Conclusion & Recommendations

**Summary:**
* The Random Forest model successfully predicts house prices with an $R^2$ score of **0.8919**.
* **Key Drivers:** The feature importance plot highlights that **OverallQual** and **1stFlrSF** are the strongest indicators of a house's value.
* **Model Accuracy:** On average, our predictions are off by approximately **$19,746.07**.

**Recommendations:**
1.  **Focus on Quality:** Since `OverallQual` is the most significant predictor, renovations that improve the overall finish of the home will yield the highest return on investment.
2.  **Size Matters:** `1stFlrSF` and `2ndFlrSF` (Living area) are critical. Extensions or finishing basements to add living space is highly valuable.

# Part 2 : improve the model by selecting the best features

```python
# Calculate correlation of ALL numerical columns with SalePrice
numeric_df = df.select_dtypes(include=[np.number]) # Get only number columns
correlations = numeric_df.corr()['SalePrice'].sort_values(ascending=False)

# Print top 10 and bottom 5 correlations
print("Top 10 Positively Correlated Features:")
print(correlations.head(15))
print("\nTop 5 Negatively Correlated Features:")
print(correlations.tail(5))
```

### 9. Feature Expansion & Engineering
To improve the model, we will:
1.  **Add Categorical Features:** Location (`Neighborhood`) and `KitchenQual` are critical predictors that we previously ignored.
2.  **One-Hot Encoding:** Random Forest cannot read text like "OldTown". We convert these into numeric columns (e.g., `Neighborhood_OldTown`: 1 or 0).
3.  **Create New Train/Val Splits:** We need to update our training data to include these new features.

```python
# 1. Identify categorical columns (Text data)
# included the features with correlation above 0.5
cat_features = ['Neighborhood', 'KitchenQual', 'CentralAir']
num_features = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd' , 'YearBuilt', 'YearRemodAdd']

# 2. Combine lists
all_features = num_features + cat_features

X_enhanced = df[all_features].copy()
y_enhanced = df['SalePrice']

# 3. Preprocessing using One-Hot Encoding
# This turns "Neighborhood" into columns like "Neighborhood_CollgCr": 1 (Yes) or 0 (No)
X_enhanced = pd.get_dummies(X_enhanced)

# 4. Handle Missing Values (Imputation)
# We must impute AFTER encoding to ensure shape consistency
my_imputer = SimpleImputer(strategy='median')
X_final = pd.DataFrame(my_imputer.fit_transform(X_enhanced), columns=X_enhanced.columns)

# 5. Split the NEW data
# We use a new variable name (train_X_opt) to distinguish from the first model
train_X_opt, val_X_opt, train_y_opt, val_y_opt = train_test_split(X_final, y_enhanced, test_size=0.2, random_state=1)

print(f"New Feature Count: {X_final.shape[1]}") 
```

### 10. Hyperparameter Tuning (Optimization)
We will use **GridSearchCV** to find the best settings for the Random Forest. 
* **n_estimators:** Number of trees.
* **max_depth:** Limits how deep the trees grow (prevents overfitting).

```python
from sklearn.model_selection import GridSearchCV

# 1. Define the parameter grid to search
param_grid = {
    'n_estimators': [100, 200],       # Number of trees
    'max_depth': [10, 20, None],      # How deep each tree can grow
    'min_samples_split': [2, 5],      # Minimum samples required to split a node
    'max_features': ['sqrt', 1.0]     # Number of features to consider at each split
}

# Initialize model
rf_opt = RandomForestRegressor(random_state=1)

# Set up Search
grid_search = GridSearchCV(estimator=rf_opt, param_grid=param_grid, 
                           cv=3, n_jobs=-1, verbose=1, scoring='neg_mean_absolute_error')

# Fit on the NEW optimized training data
print("Training optimized model (this may take a moment)...")
grid_search.fit(train_X_opt, train_y_opt)

# Get best model
best_rf_model = grid_search.best_estimator_

print("-" * 30)
print(f"Best Parameters: {grid_search.best_params_}")
print("-" * 30)
```

### 11. Final Evaluation: Did we improve?
We compare our **Baseline Model** (8 features, default settings) against our **Optimized Model** (Enhanced features + Tuned settings).

```python
# Predict using the new optimized model
val_predictions_opt = best_rf_model.predict(val_X_opt)

# Calculate new MAE
mae_opt = mean_absolute_error(val_y_opt, val_predictions_opt)
r2_opt = r2_score(val_y_opt, val_predictions_opt)

# Output results
print(f"Baseline MAE:  ${mae:,.2f}")
print(f"Optimized MAE: ${mae_opt:,.2f}")
print(f"Improvement:   ${(mae - mae_opt):,.2f}")

# Graph Comparison
results_df = pd.DataFrame({
    'Model': ['Baseline', 'Optimized'],
    'MAE': [mae, mae_opt]
})

plt.figure(figsize=(8, 5))
sns.barplot(x='Model', y='MAE', data=results_df, palette='magma')
plt.title('Model Comparison: Mean Absolute Error (Lower is Better)')
plt.ylabel('Error ($)')
plt.show()
```

### 12. Final Conclusion & Analysis

**The Results:**
* **Baseline MAE:** \$19,746
* **Optimized MAE:** \$17,837
* **Total Improvement:** Prediction error reduced by **\$1,909** per house.

**What does this mean?**
By adding categorical features (like `Neighborhood` and `KitchenQual`) and tuning the model's "brain" (Hyperparameters), we improved accuracy by approximately **9.6%**.

**Why did the model improve?**
1.  **Location, Location, Location:** Including `Neighborhood` gave the model critical context. A small house in a wealthy neighborhood is worth more than a large house in a cheaper area. The baseline model didn't know this; the optimized model does.
2.  **Kitchen Quality:** In real estate, kitchens sell houses. Adding `KitchenQual` allowed the model to differentiate between a standard home and a luxury one, even if they were the same size.
3.  **Tuning:** The `GridSearchCV` found the "sweet spot" for model complexity. It likely found that using more trees (`n_estimators=200`) helped average out errors better than the default 100.

**Business Impact:**
If this model were used by a real estate firm to estimate portfolio value for 1,000 homes, this optimization would save them **\$1.9 million** in cumulative estimation error ($1,909 \times 1,000$).
